```python
from upc import showG
```

```python
f = open('2.in')
def input():
    return f.readline()

n,m,k = map(int, input().split())
nodes = [[] for _ in range(n)]
grade = [0]*n

for _ in range(m):
    a, b = map(int, input().split())
    a-=1
    b-=1
    nodes[a].append(b)
    nodes[b].append(a)
    grade[a] += 1
    grade[b] += 1

print(grade)
showG(nodes)
```

```python
q = []
for i in range(n):
    if grade[i] < k:
        q.append(i)

while len(q):
    temp = q.pop()
    grade[temp] = 0
    for e in nodes[temp]:
        if grade[e] != 0:
            grade[e] -= 1
            nodes[e].remove(temp)
            if grade[e] < k:
                q.append(e)
    nodes[temp] = []

print(grade)
print(nodes)
showG(nodes)
```
